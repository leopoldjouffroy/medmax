```python
import numpy as np
import matplotlib.pyplot as plt

# Homemade
from ProjectiveTransformation import ProjectiveTransformation
```

```python
# Match a rectangle to a trapezoid

# rectangle (source)
source = np.array([[0,2,0,2],[0,0,1,1],[1,1,1,1]])
# trapezoid (target)
target = np.array([[0,2,0,2],[0,0,1,2],[1,1,1,1]])

#display
plt.plot(source[0,:],source[1,:],'ro')
plt.plot(target[0,:],target[1,:],'b+')
plt.show()
```

```python
result = ProjectiveTransformation(source,target)
print(result)

# display
plt.plot(result[0,:],result[1,:],'go')
plt.plot(source[0,:],source[1,:],'r+')
plt.plot(target[0,:],target[1,:],'b+')
plt.show()
```

```python
# match a random point cloud to the same cloud, but translated and distorded

n = 100

# source
source = np.random.randn(2,n)

#target
M = np.array([[3,1],[1,3]])
t = np.ones((2,1))
target = np.dot(M,source)+t

# homogeneous coordinates
source = np.vstack((source,np.ones((1,n))))
target = np.vstack((target,np.ones((1,n))))

#display
plt.plot(source[0,:],source[1,:],'ro')
plt.plot(target[0,:],target[1,:],'b+')
plt.show()
```

```python
result = ProjectiveTransformation(source,target)

# display
plt.plot(result[0,:],result[1,:],'go')
plt.plot(source[0,:],source[1,:],'r+')
plt.plot(target[0,:],target[1,:],'b+')
plt.show()
```
